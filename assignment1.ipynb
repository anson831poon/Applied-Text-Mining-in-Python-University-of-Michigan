```python
import pandas as pd

doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

df = pd.Series(doc)
df.head(10)
```

```python
df_500=df.str.extract(r'(?P<MMDDYYYY>\d{1,2}[\/-]\d{1,2}[\/-](?:19|20)\d{2})|(?P<MMYYYY>\d{1,2}[\/-](?:19|20)\d{2})|(?P<DDMMMYYYY>\d{2}\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*\s\d{2,4})|(?P<MMMDDYYYY>(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*\W+\d{1,2}\W+\d{4})|(?P<MMMYYYY>(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*\W+\d{4})|(?P<MMDDYY>\d{1,2}[\/-]\d{1,2}[\/-]\d{2}\b)|(?P<YYYY>(?:19|20)\d{2})')
df_500
```

```python
df_500['MMDDYYYY'].str.extractall(r'(?P<Month>\d{1,2})/(?P<Date>\d{1,2})/(?P<Year>\d{4})')
```

```python
df_500['MMYYYY'].str.extractall(r'(?P<Month>\d{1,2})/(?P<Year>\d{4})')
```

```python
df_500['DDMMMYYYY'].str.extractall(r'(?P<Date>\d{1,2}) (?P<Month>\w+) (?P<Year>\d{4})')
```

```python
df_500['MMMDDYYYY'].str.extractall(r'(?P<Month>\w+)\W+(?P<Date>\d{1,2})\W+(?P<Year>\d{4})')
```

```python
df_500['MMMYYYY'].str.extractall(r'(?P<Month>\w+)\W+(?P<Year>\d{4})')
```

```python
df_500['MMDDYY'].str.extractall(r'(?P<Month>\d{1,2})[/-](?P<Date>\d{1,2})[/-](?P<Year>\d{2})')
```

```python
df_500['YYYY'].str.extractall(r'(?P<Year>\d{4})')
```

```python
df_500_intermediate=pd.concat([df_500['MMDDYYYY'].str.extractall(r'(?P<Month>\d{1,2})/(?P<Date>\d{1,2})/(?P<Year>\d{4})'),
          df_500['MMDDYY'].str.extractall(r'(?P<Month>\d{1,2})[/-](?P<Date>\d{1,2})[/-](?P<Year>\d{2})'),
          df_500['DDMMMYYYY'].str.extractall(r'(?P<Date>\d{1,2}) (?P<Month>\w+) (?P<Year>\d{4})'),
          df_500['MMMDDYYYY'].str.extractall(r'(?P<Month>\w+)\W+(?P<Date>\d{1,2})\W+(?P<Year>\d{4})'),
          df_500['MMMYYYY'].str.extractall(r'(?P<Month>\w+)\W+(?P<Year>\d{4})'),
          df_500['MMYYYY'].str.extractall(r'(?P<Month>\d{1,2})/(?P<Year>\d{4})'),
          df_500['YYYY'].str.extractall(r'(?P<Year>\d{4})')],axis=0)
df_500_intermediate
```

```python
df_500_intermediate['Year']=df_500_intermediate['Year'].apply(lambda x: int('19'+str(x)) if len(x)==2 else str(x).strip())
df_500_intermediate['Year']
```

```python
df_500_intermediate['Date']=df_500_intermediate['Date'].fillna(1)
```

```python
df_500_intermediate['Date']=df_500_intermediate['Date'].apply(lambda x: x[1:] if str(x).startswith('0') else str(x).strip())
df_500_intermediate['Date']
```

```python
df_500_intermediate['Month']=df_500_intermediate['Month'].fillna(1)
```

```python
df_500_intermediate['Month']=df_500_intermediate['Month'].apply(lambda x: x[1:] if str(x).startswith('0') else 1 if str(x).startswith('Jan')
                                                                 else 2 if str(x).startswith('Feb') else 3 if str(x).startswith('Mar')
                                                                 else 4 if str(x).startswith('Apr') else 5 if str(x).startswith('May')
                                                                 else 6 if str(x).startswith('Jun') else 7 if str(x).startswith('Jul')
                                                                 else 8 if str(x).startswith('Aug') else 9 if str(x).startswith('Sep')
                                                                 else 10 if str(x).startswith('Oct') else 11 if str(x).startswith('Nov')
                                                                 else 12 if str(x).startswith('Dec') else str(x).strip())
df_500_intermediate['Month']
```

```python
df_500_intermediate['Date']=df_500_intermediate['Date'].astype(int)
df_500_intermediate['Month']=df_500_intermediate['Month'].astype(int)
df_500_intermediate['Year']=df_500_intermediate['Year'].astype(int)
```

```python
df_500_final=df_500_intermediate.sort_values(['Year','Month','Date'])[['Year','Month','Date']].droplevel(1)
df_500_final
```
